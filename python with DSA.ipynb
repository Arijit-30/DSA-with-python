```python
#strings
def count_char(s, ch):
    c = 0
    for i in s:
        if i == ch:
            c = c + 1
    return c

s = input()
ch = input()
print(count_char(s, ch))
```

```python
def replace_vowel(s):
    v = "aeiouAEIOU"
    r = ""
    for i in s:
        if i in v:
            r = r + "*"
        else:
            r = r + i
    return r

s = input()
print(replace_vowel(s))
```

```python
s = list(input())
i = 0
j = len(s) - 1
while i < j:
    s[i], s[j] = s[j], s[i]
    i = i + 1
    j = j - 1
print("".join(s))
```

```python
def reverse_new(s):
    r = ""
    i = len(s) - 1
    while i >= 0:
        r = r + s[i]
        i = i - 1
    return r

s = input()
print(reverse_new(s))
```

```python
def check_pal(s):
    if s == s[::-1]:
        return "Palindrome"
    else:
        return "Not Palindrome"

s = input()
print(check_pal(s))
```

```python
def title_case(s):
    w = s.split()
    r = ""
    for i in w:
        r = r + i.capitalize() + " "
    return r.strip()

s = input()
print(title_case(s))
def deleteChar(s, ch):
    r = ""
    for i in s:
        if i != ch:
            r = r + i
    return r

s = input()
ch = input()
print(deleteChar(s, ch))
```

```python
def sum_digits(s):
    total = 0
    for i in s:
        if i.isdigit():
            total = total + int(i)
    return total

s = input()
print(sum_digits(s))
```

```python
#lists
#increments elements by 5
def inc_list(a):
    for i in range(len(a)):
        a[i] = a[i] + 5
    return a

a = list(map(int, input().split()))
print(inc_list(a))
```

```python
#count occurence ofelement
def count_ele(a, x):
    c = 0
    for i in a:
        if i == x:
            c = c + 1
    return c

a = list(map(int, input().split()))
x = int(input())
print(count_ele(a, x))
```

```python
#larget elemtn 
def largest(a):
    m = a[0]
    for i in a:
        if i > m:
            m = i
    return m

a = list(map(int, input().split()))
print(largest(a))
```

```python
#secodn largest 
def second_largest(a):
    a = list(set(a))
    a.sort()
    return a[-2]

a = list(map(int, input().split()))
print(second_largest(a))
```

```python
#median list
def median(a):
    a.sort()
    n = len(a)
    if n % 2 == 0:
        return (a[n//2] + a[n//2 - 1]) / 2
    else:
        return a[n//2]

a = list(map(int, input().split()))
print(median(a))
```

```python
def binary_search(arr, low, high, target):
    if low > high:
        return -1

    mid = (low + high) // 2

    if arr[mid] == target:
        return mid

    elif target < arr[mid]:
        return binary_search(arr, low, mid - 1, target)

    else:
        return binary_search(arr, mid + 1, high, target)


arr = [2, 5, 8, 12, 16, 23, 38]
target = 16
```
